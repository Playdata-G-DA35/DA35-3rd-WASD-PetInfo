```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
print(wordnet.NOUN, wordnet.ADJ, wordnet.VERB, wordnet.ADV) # 명사, 형용사, 동사, 부사
```

```python
# Pos-tag 에서 반환한 품사표기(펜 트리뱅크 태그세트)을 WordNetLemmatizer의 품사표기로 변환
def get_wordnet_pos(pos_tag):
    """
    펜 트리뱅크 품사표기를 WordNetLemmatizer에서 사용하는 품사표기로 변환
    형용사/동사/명사/부사 표기 변환
    """
    if pos_tag.startswith("J"):
        return wordnet.ADJ
    elif pos_tag.startswith("V"):
        return wordnet.VERB
    elif pos_tag.startswith("N"):
        return wordnet.NOUN
    elif pos_tag.startswith("R"):
        return wordnet.ADV
    else:
        return None
```

```python
def tokenize_text(text):
    """text 토큰화 처리 함수"""
    #1.소문자 변환
    text = text.lower()
    #2. 문장단위로 토큰화
    sent_tokens = nltk.sent_tokenize(text) #[문장1, 문장2, 문장3,..]

    ###토큰화 + cleaning작업
    #3. stopword들 loading
    stop_words = stopwords.words("english") # 필요하다면 더 추가, 제거할 수도 있다.
    stop_words.extend(["although", "unless", "may", ",", "'s", "'", ".", "\n"])
    #4. lemmatizer객체 생성
    lemm = WordNetLemmatizer()

    #최종 결과를 담을 리스트
    result_tokens = []
    #문장별로 처리
    for sent in sent_tokens:
        #regexp 토큰화
        word_tokens = nltk.regexp_tokenize(sent, r"[a-zA-Z]+") #[단어1, 단어2, 단어3, ..]
        #불용어 제거
        word_tokens = [word for word in word_tokens if word not in stop_words]

        #원형 복원
        ##품사부탁
        word_tokens = pos_tag(word_tokens) #[(단어1, 품사), (단어2, 품사), (단어3, 품사),..]
        ##lemmatizer에서 사용하는 품사 string으로 변환.: "NN" -> "n"
        word_tokens = [ (word, get_wordnet_pos(pos)) \
                       for word, pos in word_tokens \
                       if get_wordnet_pos(pos) is not None]

        ##원형복원
        word_tokens = [lemm.lemmatize(word, pos=pos) for word, pos in word_tokens]
        result_tokens.append(word_tokens)

    return result_tokens
```

```python
b_info_list #문장리스트
```

```python
tok_result = []
for word in b_info_list:
    tok_result += tokenize_text(str(word))
tok_result
```

```python
# 중첩되어있는 2차원 리스트를 1차원으로 바꿔야함
dim_one_words = []
for lst in tok_result:
    dim_one_words += lst
dim_one_words
```

```python
from nltk import Text
from nltk import FreqDist
import matplotlib.pyplot as plt
```

```python
pet_text = Text(dim_one_words, name="반려견") #임의로 이름을 정해줌 name="반려견"
# print(dim_one_words)
pd_dim = pd.DataFrame(dim_one_words)
# pd_dim
as_sort = pd_dim.groupby([0]).size().sort_values(ascending=False).copy()
copy_as = as_sort.iloc[:20].copy()
plt.figure(figsize=(12,4))
plt.plot(copy_as.index, copy_as.values, marker="o", linestyle="-", color="b")
plt.xlabel("Index")
plt.ylabel("Values")
plt.title("Line Plot: Series Index vs. Values")
plt.show()
# as_sort.iloc[:, 0]
# as_sort.loc[0]
# copy_as = as_sort.iloc[:20].copy()
# copy_as[:20]
```

```python
plt.figure(figsize=(10,4))
plt.title("word freq")
pet_text.plot(20) #빈도수가 큰 상위20개 단어에 대해 선그래프를 작성
plt.show()
```

```python
##특정 단어가 문서의 어느부분들에 나오는지 시각화
plt.figure(figsize=(10,10))
pet_text.dispersion_plot(
    ["dog", "coat", "breed", "inch", "terrier"]
)
plt.show()
```

```python
fd = pet_text.vocab()
print("총 토큰수: ", fd.N())
print("고유 토큰의 개수: ", fd.B())
word = "dog"
print(f"{word}의 빈도수: {fd.get(word)}")
print(f"{word}의 비율: {fd.freq(word)}")
fd.most_common(10) #상위 10개 보기
```

```python
#WordCloud 객체 생성 -> 어떻게 그릴지 설정
wc = WordCloud(
    max_words = 50, # 최대 몇개단어를 사용할지설정. 빈도수 많은 순서대로
    prefer_horizontal=0.7, #가로쓰기 비율(기본:0.9)
    width=500, #가로크기:픽셀
    height=500, #세로크기:픽셀
    relative_scaling=0.5, #빈도수가 증가할 때 글씨 크기를 얼마나 크게 할지 비율
                          #(단계적 크기: 1이면 두배)
    min_font_size=30,
    max_font_size=100,
    background_color="white"
)
```

```python
# generate_from_frequencies({"단어":빈도}) => 딕셔너리형태 fd이 딕셔너리 형태이다
word_cloud = wc.generate_from_frequencies(fd) # fd는 딕셔너리
#파일로 저장
word_cloud.to_file("pet_info_wordcloud.png")
```
